```python
import tensorflow as tf
from tensorflow.keras.layers import Input, Dense, Dropout, BatchNormalization, Add
from tensorflow.keras.models import Model
from tensorflow.keras.regularizers import l2
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau
from tensorflow.keras.optimizers import Adam
from sklearn.model_selection import StratifiedKFold
from sklearn.metrics import f1_score
import numpy as np

def create_model(input_shape):
    input_tensor = Input(shape=(input_shape,))
    x = Dense(64, activation='relu', kernel_regularizer=l2(0.01))(input_tensor)
    x = BatchNormalization()(x)
    x = residual_block(x, 64, dropout_rate=0.5, l2_reg=0.01)
    x = residual_block(x, 64, dropout_rate=0.5, l2_reg=0.01)
    output_tensor = Dense(1, activation='sigmoid')(x)
    model = Model(inputs=input_tensor, outputs=output_tensor)
    model.compile(optimizer=Adam(learning_rate=0.00005), loss='binary_crossentropy', metrics=['accuracy', 'AUC'])
    return model

def residual_block(x, units, dropout_rate=0.3, l2_reg=0.001):
    shortcut = x
    x = Dense(units, activation='relu', kernel_regularizer=l2(l2_reg))(x)
    x = BatchNormalization()(x)
    x = Dropout(dropout_rate)(x)
    x = Dense(units, activation=None, kernel_regularizer=l2(l2_reg))(x)
    x = BatchNormalization()(x)
    x = Add()([shortcut, x])
    x = tf.keras.activations.relu(x)
    return x

def optimize_threshold(y_true, y_probs):
    thresholds = np.arange(0.1, 0.9, 0.01)
    f1_scores = [f1_score(y_true, (y_probs > t).astype(int)) for t in thresholds]
    return thresholds[np.argmax(f1_scores)]

# Cross-validation com otimização de F1 Score
kf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
f1_scores = []
best_thresholds = []

for train_index, val_index in kf.split(X_train, y_train):
    X_tr, X_val = X_train.iloc[train_index], X_train.iloc[val_index]
    y_tr, y_val = y_train.iloc[train_index], y_train.iloc[val_index]

    # Cria e treina um novo modelo para cada fold
    model = create_model(X_train.shape[1])
    history = model.fit(X_tr, y_tr, validation_data=(X_val, y_val), epochs=100, batch_size=32,
                        callbacks=[EarlyStopping(monitor='val_loss', patience=15, restore_best_weights=True),
                                   ReduceLROnPlateau(monitor='val_loss', factor=0.2, patience=7, min_lr=1e-7)],
                        verbose=0)

    # Previsões no conjunto de validação
    y_val_probs = model.predict(X_val)
    best_threshold = optimize_threshold(y_val, y_val_probs)
    y_val_pred = (y_val_probs > best_threshold).astype(int)

    f1 = f1_score(y_val, y_val_pred)
    f1_scores.append(f1)
    best_thresholds.append(best_threshold)

    print(f'Fold F1 Score: {f1:.4f} with threshold: {best_threshold:.2f}')

# Avaliação final no conjunto de teste
model = create_model(X_train.shape[1])
model.fit(X_train, y_train, epochs=100, batch_size=32,
          callbacks=[EarlyStopping(monitor='val_loss', patience=15, restore_best_weights=True),
                     ReduceLROnPlateau(monitor='val_loss', factor=0.2, patience=7, min_lr=1e-7)],
          verbose=0)

# Usando o melhor threshold médio para o conjunto de teste
best_threshold = np.mean(best_thresholds)
y_test_probs = model.predict(X_test)
y_test_pred = (y_test_probs > best_threshold).astype(int)
final_f1 = f1_score(y_test, y_test_pred)

print(f'Mean F1 Score from CV: {np.mean(f1_scores):.4f} ± {np.std(f1_scores):.4f}')
print(f'Test F1 Score: {final_f1:.4f} with threshold: {best_threshold:.2f}')
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau, Callback
from tensorflow.keras.regularizers import l2
from sklearn.metrics import f1_score, precision_recall_curve
import numpy as np

# Callback personalizado para calcular F1 score e encontrar o melhor threshold
class F1ScoreCallback(Callback):
    def __init__(self, validation_data):
        super(F1ScoreCallback, self).__init__()
        self.validation_data = validation_data
        self.best_f1 = 0
        self.best_threshold = 0

    def on_epoch_end(self, epoch, logs=None):
        X_val, y_val = self.validation_data
        y_pred_probs = self.model.predict(X_val)
        precision, recall, thresholds = precision_recall_curve(y_val, y_pred_probs)
        f1_scores = 2 * (precision * recall) / (precision + recall)
        optimal_idx = np.argmax(f1_scores)
        current_f1 = f1_scores[optimal_idx]
        current_threshold = thresholds[optimal_idx] if optimal_idx < len(thresholds) else 1.0
        if current_f1 > self.best_f1:
            self.best_f1 = current_f1
            self.best_threshold = current_threshold
            print(f"Epoch {epoch+1}: Improved F1 to {self.best_f1:.4f} with threshold {self.best_threshold:.4f}")
        else:
            print(f"Epoch {epoch+1}: F1 {current_f1:.4f} with threshold {current_threshold:.4f}")

# Definição do modelo
def create_model():
    model = Sequential([
        Dense(16, input_shape=(X_train.shape[1],), activation='relu', kernel_regularizer='l2'),
        Dropout(0.2),
        Dense(32, activation='relu', kernel_regularizer='l2'),
        Dropout(0.2),
        Dense(64, activation='relu', kernel_regularizer='l2'),
        Dropout(0.2),
        Dense(128, activation='relu', kernel_regularizer='l2'),
        Dropout(0.2),
        Dense(64, activation='relu', kernel_regularizer='l2'),
        Dropout(0.2),
        Dense(32, activation='relu', kernel_regularizer='l2'),
        Dropout(0.2),
        Dense(16, activation='relu', kernel_regularizer='l2'),
        Dropout(0.2),
        Dense(1, activation='sigmoid')
    ])
    model.compile(optimizer=Adam(learning_rate=0.001), loss='binary_crossentropy', metrics=['accuracy'])
    return model

# Treinamento e validação
model = create_model()
f1_callback = F1ScoreCallback(validation_data=(X_val, y_val))
early_stopping = EarlyStopping(monitor='val_loss', patience=50, restore_best_weights=True)
reduce_lr = ReduceLROnPlateau(monitor='val_loss', factor=0.2, patience=5, min_lr=0.0001, verbose=1)

model.fit(X_train, y_train, validation_data=(X_val, y_val), epochs=100, batch_size=32,
          callbacks=[early_stopping, reduce_lr, f1_callback])

# Avaliação usando o melhor threshold encontrado
y_test_probs = model.predict(X_test)
y_test_pred = (y_test_probs > f1_callback.best_threshold).astype(int)
final_f1 = f1_score(y_test, y_test_pred)

print(f'Final Test F1 Score: {final_f1:.4f} with threshold: {f1_callback.best_threshold:.4f}')
```

```python
import numpy as np
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier, VotingClassifier
from sklearn.metrics import accuracy_score, f1_score
from sklearn.model_selection import cross_val_score

# Primeiro nível da cascata: MLP (best_model)
with tf.device('/device:GPU:0'):
    best_model.fit(X_train, y_train)

# Previsões do primeiro nível
with tf.device('/device:GPU:0'):
    mlp_prob = best_model.predict(X_test)  # Probabilidades (saída da camada sigmoid)
    mlp_pred = (mlp_prob > 0.5).astype(int)  # Convertendo probabilidades em classes binárias

# Avaliação geral do primeiro nível
mlp_acc = accuracy_score(y_test, mlp_pred)
mlp_f1 = f1_score(y_test, mlp_pred, average='binary')
print(f'Accuracy do MLP: {mlp_acc:.4f}')
print(f'F1 Score do MLP: {mlp_f1:.4f}')

# Preparando para o segundo nível da cascata: Ensemble Learning
nb_model = GaussianNB()
svm_model = SVC(probability=True)
rf_model = RandomForestClassifier()

ensemble_model = VotingClassifier(estimators=[
    ('nb', nb_model),
    ('svm', svm_model),
    ('rf', rf_model)],
    voting='soft')

ensemble_model.fit(X_train, y_train)

ensemble_pred = ensemble_model.predict(X_test)

# Avaliação geral do segundo nível
ensemble_acc = accuracy_score(y_test, ensemble_pred)
ensemble_f1 = f1_score(y_test, ensemble_pred, average='binary')
print(f'Accuracy do Ensemble Learning: {ensemble_acc:.4f}')
print(f'F1 Score do Ensemble Learning: {ensemble_f1:.4f}')

# Validação cruzada no conjunto total para verificar a robustez do ensemble
scores = cross_val_score(ensemble_model, X_train, y_train, cv=5, scoring='f1_weighted')
print(f'F1 Score Médio (Cross-Validation no Ensemble): {scores.mean():.4f}')

```
